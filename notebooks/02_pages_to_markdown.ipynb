# CECCE Chatbot - Phase 2

## Pages to Markdown

Dans cette étape, je vais récupérer les pages HTML du Google Sites et convertir leur contenu en Markdown.

L’objectif est de nettoyer les pages afin de conserver uniquement l’information importante qui sera utilisée plus tard par Gemini File Search.

## Objectifs

- Télécharger les pages HTML
- Nettoyer le contenu
- Retirer les éléments inutiles
- Convertir les pages en Markdown
- Sauvegarder les fichiers dans le dossier markdown/

```python
import os
import requests
import pandas as pd

# créer le dossier raw_pages
os.makedirs("../data/raw_pages", exist_ok=True)

# lire le fichier sitemap
df = pd.read_csv("../data/manifests/site_map.csv")

print("Nombre de pages trouvées :", len(df))

# télécharger les pages
for index, row in df.iterrows():

    url = row["url"]

    try:

        response = requests.get(url, timeout=10)

        if response.status_code == 200:

            filename = f"page_{index}.html"

            filepath = os.path.join(
                "../data/raw_pages",
                filename
            )

            with open(filepath, "w", encoding="utf-8") as file:
                file.write(response.text)

            print(f"Téléchargé : {filename}")

        else:

            print(f"Erreur {response.status_code} : {url}")

    except Exception as e:

        print(f"Erreur : {e}")

print("\nTéléchargement HTML terminé.")
```

## Prochaine étape

La prochaine phase du projet consistera à récupérer et sauvegarder les documents PDF présents dans le Google Sites afin d’élargir le corpus informationnel utilisé par le chatbot.

```python
import os
from bs4 import BeautifulSoup
from markdownify import markdownify as md

# créer dossier markdown
os.makedirs("../data/markdown", exist_ok=True)

raw_pages_folder = "../data/raw_pages"

html_files = os.listdir(raw_pages_folder)

print("Nombre de fichiers HTML :", len(html_files))

for filename in html_files:

    if filename.endswith(".html"):

        filepath = os.path.join(raw_pages_folder, filename)

        with open(filepath, "r", encoding="utf-8") as file:
            html_content = file.read()

        soup = BeautifulSoup(html_content, "html.parser")

        markdown_content = md(str(soup))

        markdown_filename = filename.replace(".html", ".md")

        markdown_path = os.path.join(
            "../data/markdown",
            markdown_filename
        )

        with open(markdown_path, "w", encoding="utf-8") as md_file:
            md_file.write(markdown_content)

        print(f"Converti : {markdown_filename}")

print("\nConversion Markdown terminée.")
```
