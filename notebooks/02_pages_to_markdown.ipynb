# CECCE Chatbot - Phase 2

## Pages to Markdown

Dans cette étape, je vais récupérer les pages HTML du Google Sites et convertir leur contenu en Markdown.

L’objectif est de nettoyer les pages afin de conserver uniquement l’information importante qui sera utilisée plus tard par Gemini File Search.

## Objectifs

- Télécharger les pages HTML
- Nettoyer le contenu
- Retirer les éléments inutiles
- Convertir les pages en Markdown
- Sauvegarder les fichiers dans le dossier markdown/
