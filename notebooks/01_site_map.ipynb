# Prochaines étapes

Après cette phase, je vais :

1. Télécharger les pages HTML
2. Convertir les pages en Markdown
3. Télécharger les documents PDF
4. Préparer les métadonnées
5. Envoyer les fichiers dans Gemini File Search

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

START_URL = "https://sites.google.com/ecolecatholique.ca/mourad/accueil"

response = requests.get(START_URL)

print("Status Code:", response.status_code)

soup = BeautifulSoup(response.text, "html.parser")

links = set()

for a_tag in soup.find_all("a", href=True):
    href = a_tag["href"]

    full_url = urljoin(START_URL, href)

    if "ecolecatholique.ca" in full_url:
        links.add(full_url)

print("\nPages trouvées :\n")

for link in sorted(links):
    print(link)
```

```python
import pandas as pd

clean_links = []

for link in sorted(links):

    # garder seulement les pages Google Sites
    if "sites.google.com" in link:

        # retirer les liens inutiles
        if "#" not in link and "javascript:" not in link:

            clean_links.append(link)

# retirer les doublons
clean_links = list(set(clean_links))

print(f"Nombre de pages trouvées : {len(clean_links)}")

print("\nPages retenues :\n")

for page in clean_links:
    print(page)

# créer dataframe
df = pd.DataFrame(clean_links, columns=["url"])

# sauvegarder le manifest
df.to_csv("../data/manifests/site_map.csv", index=False)

print("\nsite_map.csv sauvegardé dans data/manifests/")
```
