# CECCE Chatbot - Phase 0

## Project Bootstrap and Evaluation Set

Dans cette première phase, je vais préparer la base du projet afin de faciliter les prochaines étapes du développement du chatbot.

## Objectifs de la Phase 0

- Préparer la structure du projet
- Définir le périmètre du POC
- Préparer les questions d’évaluation
- Configurer l’environnement Python
- Préparer les notebooks du projet
- Configurer Gemini API

## URL principale du Google Sites

https://sites.google.com/ecolecatholique.ca/mourad/accueil

## Scope du projet

Pour la première version du projet, je vais utiliser les pages publiques du Google Sites CECCE ainsi que certains documents PDF accessibles publiquement.

Je vais commencer avec un sous-ensemble limité de pages importantes afin de valider le comportement du chatbot avant d’élargir le corpus complet.

## Résultat attendu

À la fin de cette phase, le projet devrait être organisé, documenté et prêt pour commencer le crawl du Google Sites.
