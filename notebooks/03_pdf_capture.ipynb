# PDF Capture

Dans cette phase, je vais détecter et télécharger les documents PDF présents dans le Google Sites public du CECCE.

Ces documents vont faire partie du corpus informationnel utilisé plus tard par Gemini File Search et le chatbot.

---

# Objectifs de la Phase 3

- Identifier les liens PDF présents dans le site
- Télécharger les documents PDF
- Sauvegarder les fichiers localement
- Préparer les documents pour l’indexation
- Élargir le corpus informationnel du chatbot

---

# Résultat attendu

À la fin de cette phase, je veux obtenir une collection de documents PDF sauvegardés dans le projet afin de préparer les prochaines étapes d’indexation et de recherche documentaire.

```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# créer dossier PDF
os.makedirs("../data/raw_pdfs", exist_ok=True)

# lire sitemap
df = pd.read_csv("../data/manifests/site_map.csv")

pdf_links = []

print("Recherche des PDF...\n")

for index, row in df.iterrows():

    url = row["url"]

    try:

        response = requests.get(url, timeout=10)

        soup = BeautifulSoup(response.text, "html.parser")

        for a_tag in soup.find_all("a", href=True):

            href = a_tag["href"]

            full_url = urljoin(url, href)

            if ".pdf" in full_url.lower():

                pdf_links.append(full_url)

    except Exception as e:

        print(f"Erreur : {e}")

# retirer doublons
pdf_links = list(set(pdf_links))

print(f"\nNombre de PDF trouvés : {len(pdf_links)}\n")

# télécharger PDF
for index, pdf_url in enumerate(pdf_links):

    try:

        response = requests.get(pdf_url, timeout=15)

        if response.status_code == 200:

            filename = f"document_{index}.pdf"

            filepath = os.path.join(
                "../data/raw_pdfs",
                filename
            )

            with open(filepath, "wb") as file:
                file.write(response.content)

            print(f"Téléchargé : {filename}")

    except Exception as e:

        print(f"Erreur téléchargement PDF : {e}")

print("\nTéléchargement PDF terminé.")
```
